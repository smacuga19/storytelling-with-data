## Problem 2: word frequency counter

Given a string of text, count how often each word appears. This is the foundation of search engines, spam filters, and basic text analysis.

```python
text = "the quick brown fox jumps over the lazy dog the fox is quick"

# Step 1: split the text into individual words
words = text.lower().split()
print('Words:', words)

# Step 2: count each word
counts = {}
for word in words:
    counts[word] = counts.get(word, 0) + 1
print('Counts:', counts)

# Step 3: sort by frequency (descending)
sorted_words = sorted(counts.items(), key=lambda x: -x[1])
for word, count in sorted_words:
    print(f'{word}: {count}')
```

## Problem 3: reading a CSV without a library

Next week we'll use Pandas, which reads CSVs in one line. But understanding what's actually happening helps you debug when Pandas does something unexpected.

```python
# First, create a small CSV file to practice with
csv_text = '''name,age,major
Ada,22,Math
Alan,19,Logic
Grace,21,CS
Donald,20,Physics'''

with open('students.csv', 'w') as f:
    f.write(csv_text)

print('CSV created!')
```

```python
# Now read it back manually
with open('students.csv', 'r') as f:
    lines = f.readlines()

header = lines[0].strip().split(',')
rows = []
for line in lines[1:]:
    values = line.strip().split(',')
    row = dict(zip(header, values))
    rows.append(row)

# rows is now a list of dicts — one per student
for row in rows:
    print(row)
```

## Problem 4: debugging practice

The function below is supposed to return the average of a list of numbers. It has a bug. Find it, fix it, and add a test that would have caught the bug.

```python
def average(numbers):
    total = 0
    for i in range(len(numbers)):
        total = numbers[i]
    return total / len(numbers)

# Does this look right?
print(average([1, 2, 3]))       # expected: 2.0
print(average([10, 20, 30, 40])) # expected: 25.0
```

**Debugging strategy:**

1. Read the code out loud — what does each line *actually* do?
2. Trace through with a small example on paper
3. Add `print` statements to show intermediate values
4. Ask AI: *"what does this function do, step by step?"*

Once you find the bug, fix it in the cell below and confirm both tests pass.

```python
def average(numbers):
    # FIXED version
    total = 0
    for i in range(len(numbers)):
        total += numbers[i]   # += instead of =
    return total / len(numbers)

print(average([1, 2, 3]))        # should print 2.0
print(average([10, 20, 30, 40])) # should print 25.0
```

## Problem 5: a tiny simulation

Simulate rolling two dice 10,000 times and count how often each sum (from 2 to 12) appears. Then print the results.

This problem combines loops, conditionals, dictionaries, and the `random` module. It also previews data analysis: you're generating data and summarizing it.

```python
import random

n_rolls = 10000
sum_counts = {s: 0 for s in range(2, 13)}

for _ in range(n_rolls):
    die1 = random.randint(1, 6)
    die2 = random.randint(1, 6)
    total = die1 + die2
    sum_counts[total] += 1

# Print a little histogram
for s in range(2, 13):
    bar = '█' * (sum_counts[s] // 50)
    print(f'{s:2d}: {bar} {sum_counts[s]}')
```

**Questions to think about:**

- Which sum is most common? Why?
- Which sums are least common? Why?
- What would change if you rolled *three* dice instead of two?
- Try increasing `n_rolls` to 100,000 — how does the shape of the distribution change?
